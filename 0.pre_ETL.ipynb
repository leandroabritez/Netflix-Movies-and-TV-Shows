## Inspeccionar las primeras filas del Archivo

```python
with open(input_file, 'r', encoding='latin1') as file:
    reader = csv.reader(file)
    for i, row in enumerate(reader):
        if i < 5:  # Mostrar las primeras 5 filas
            print(row)
```

Se identifican columnas con comas adicionales, y se procede a limpiar las mismas.

## Se limpian columnas adicionales

```python
# Cargar el archivo CSV en un DataFrame de pandas
df = pd.read_csv(input_file, encoding='latin1')

# Mostrar las columnas del DataFrame
print("Columnas antes de la limpieza:")
print(df.columns)

# Eliminar columnas adicionales si son innecesarias
df_cleaned = df.iloc[:, :12]

# Guardar el DataFrame limpio en un nuevo archivo CSV
output_file = 'G:/Mi unidad/Meli/Netflix-Movies-and-TV-Shows/data/netflix_titles_cleaned.csv'
df_cleaned.to_csv(output_file, index=False, encoding='utf-8')
```

## (OPCIONAL) Alternativa para cargar los datos a SQL mediante conexión con python

```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

# Ruta al archivo CSV
file_path = 'G:/Mi unidad/Meli/Netflix-Movies-and-TV-Shows/data/netflix_titles_cleaned.csv'

# Leer el archivo CSV
df = pd.read_csv(file_path, encoding='utf-8')

# Reemplazar NaN por None para evitar errores en la inserción
df = df.where(pd.notnull(df), None)

# Conectarse a la base de datos MySQL | Modificar los datos según corresponda a tu enlace
try:
    connection = mysql.connector.connect(
        host='localhost',  
        database='netflix',
        user='root',
        password='root1234'
    )

    if connection.is_connected():
        cursor = connection.cursor()

        # Insertar los datos
        for index, row in df.iterrows():
            try:
                insert_statement = """
                INSERT INTO netflix (show_id, type_, title, director, cast, country, date_added, release_year, rating, duration, listed_in, description_)
                VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
                """
                data = (
                    row['show_id'], 
                    row['type'], 
                    row['title'], 
                    row['director'], 
                    row['cast'], 
                    row['country'], 
                    row['date_added'], 
                    row['release_year'], 
                    row['rating'], 
                    row['duration'], 
                    row['listed_in'], 
                    row['description']
                )
                cursor.execute(insert_statement, data)
            except Error as e:
                print(f"Error while inserting row {index}: {e}")
                print(data)

        # Confirmar los cambios
        connection.commit()

except Error as e:
    print(f"Error while connecting to MySQL: {e}")

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("MySQL connection is closed")
```
